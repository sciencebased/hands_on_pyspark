# KEY CONCEPT: RDDs cand hold key / value pairs

For example: number of friends by age
Key is age, value is number of friends

- Instead of just a list of ages or a list of # of friends, we can store (age, #friends)

Just maps pairs of data into the RDD. For example:

```sql
totalsByAge = rdd.map(lambda x:(x,1))
```

Spark can do special stuff with key/value data

- reduceByKey() : Combines with the same key using some function, 

rdd.reduceByKey(lambda x,y: x+y) adds them up.

- groupByKey() : Group values with the same key
- sortByKey() : Sort RDD by key values
- keys(), values(): Create an RDD of just the keys, or just the values

You can do SQL-STYLE joins on two key/value RDDs
- join, righOuterJoin, leftOuterJoin, cogroup, subtractByKey

Mapping just the values of a key/value RDD?
- With key/value data, use mapValues() and flatMapValues() if your transformation doesnt affect the keys
- It is more efficient

# Example


```sql
#Input data: ID, name, age, number of friends
0, Will, 33, 385
1, Jean-Luc, 26, 2
2, Hugh, 40, 400
3, Deanna, 26, 2
4, Quark, 68, 20
```

We will find the Average number of friends by age

- Parsing the input

```sql
def parseLine(line):
    fields = line.split(',')
    age = int(fields[2])
    numFriends = int(fields[3])
    return (age, numFriends)

lines = sc.textFile("file_location")
rdd = lines.map(parseLine)
```

Output is key/value pairs of (age, numFriends)

Result:
(33, 385)
(26, 2)
....

```sql
totalsByAge = rdd.mapValues(lambda x:(x,1)).reduceByKey(lambda x,y: (x[0]+y[0],x[1]+y[1]))
```

Compute averages

```sql
averagesByAge = totalsByAge.mapValues(lambda x: x[0]/x[1])
```

```sql
results = averagesByAge.collect()
for result in results:
    print result
```
