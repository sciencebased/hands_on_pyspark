# Transforming RDD's

You can do the following (most common):
- map
- flatmap
- filter
- distinct
- sample
- union, intersection, subtract, cartesian

BREAKDOWN:
- Map = one sort of data to another set of data (1 to 1 relationship)
- Flatmap = similar but can create multiple values (one to another, can be transformed into multiple)
- Filter = filter anything but error lines, for example
- Distinctin = Unique/Distinc, avoid duplicates, you get a smaller or exact
- Union/Intersection/Substract/Cartesian = Takes two RDDs as input and make operations over them

# Examples:

```sql
rdd = sc.parallelize([1,2,3,4]) #Harcode that list
rdd.map(lambda x: x*x)
this yields 1, 4, 9, 16
```

Notice that: Many RDD methods accept a function as a parameter, like in this case the lambda function inside map
